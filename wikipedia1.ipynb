```python
import requests

import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/Epidemiology_of_depression'
r=requests.get(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "lxml")
```

```python
print(soup)
```

```python
table=soup.find_all('table')
```

```python
print(table)
```

```python
rows_soup=soup.find_all('tr')
```

```python
rows_head=[]
for i in rows_soup:
    rows_data=[]
    for j in i.find_all(['th','td']):
        rows_data.append(j.get_text(strip=True))
    rows_head.append(rows_data)
    
    
```

```python
print(rows_head)
```

```python
df=pd.DataFrame(rows_head[1:],columns=rows_head[0])
```
